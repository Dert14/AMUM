### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
# TUTAJ UMIESC KOD
def function(x):
    result = (1/8)*pow(x, 4)+2*pow(x+0.7, 3)-(1/2)*pow(x, 2)-38*x+1
    return result
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
# TUTAJ UMIESC KOD
domain = range(-20, 21, 1)
values = [function(x) for x in domain]
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
# TUTAJ UMIESC KOD
import matplotlib.pyplot as plt


plt.plot(domain, values, label="wartości")
plt.scatter(domain, values)
plt.xlabel("os X")
plt.ylabel("os Y")
plt.title("Wykres")
plt.legend()
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
minimum_value = min(values)
minimum_in_domain = domain[values.index(minimum_value)]
print(f"value of minimum = {minimum_value:.4f}")
print(f"X of minimum = {minimum_in_domain}")
plt.plot(domain, values, label="wartości")
plt.scatter(domain, values)
plt.xlabel("os X")
plt.ylabel("os Y")
plt.title("Wykres")
plt.xlim(-5, 5)
plt.ylim(-50, 120)
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
# TUTAJ UMIESC KOD
resolution_domain = [x/10 for x in range(-200, 201, 1)]
resolution_values = [function(x) for x in resolution_domain]
minimum_resolution_value = min(resolution_values)
minimum_in_resolution_domain = resolution_domain[resolution_values.index(minimum_resolution_value)]
print(f"value of minimum {minimum_resolution_value:.4f}")
print(f"X of minimum {minimum_in_resolution_domain}")
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
# TUTAJ UMIESC KOD
plt.plot(resolution_domain, resolution_values, label="wartości")
#plt.scatter(resolution_domain, resolution_values)
plt.xlabel("os X")
plt.ylabel("os Y")
plt.title("Wykres")
plt.xlim(-5, 5)
plt.ylim(-50, 120)
plt.legend()
```
